```python
import numpy as np
import keras
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.preprocessing.text import Tokenizer
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
# numpy version 1.16.1 이후 버전부터는 allow_pickle의 디폴트 값이 False이므로 에러 발생 

np_load_old = np.load

np.load = lambda *a, **k: np_load_old(*a, allow_pickle=True, **k)
```

```python
# num_words=1000 : 빈도수가 1000번째로 많이 사용된 단어까지만 데이터셋으로 만든다, 몇번째까지인지는 직접 지정
# imdb dataset은 training dataset 25,000개 & test dataset 25,000개를 제공

(x_train, y_train),(x_test, y_test) = imdb.load_data(num_words=1000)

np.load = np_load_old
print(x_train.shape)
print(x_test.shape)
```

```python
print(x_train[0])
#print(y_train[0])
print(len(x_train[0]))
print(x_train[1])
print(len(x_train[1]))
print(x_train[100])
print(len(x_train[100]))

# 리스트의 길이는 각 리뷰(문장)의 길이(단어 사용 갯수)
# 리스트 내의 값들 : 만약 1이라면 빈도수가 가장 많은 단어가 그 위치에 사용됐다는 뜻
```

```python
tokenizer = Tokenizer(num_words=1000)
x_train = tokenizer.sequences_to_matrix(x_train, mode='binary')
x_test = tokenizer.sequences_to_matrix(x_test, mode='binary')

print(x_train[0])
```

```python
print(x_train[522])
```

```python
print(x_test[0])
```

```python
num_class = 2

y_train = keras.utils.to_categorical(y_train, num_class)
y_test = keras.utils.to_categorical(y_test, num_class)

print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()

model.add(Dense(512, activation = 'relu', input_dim=1000))
model.add(Dropout(0.5))
model.add(Dense(num_class, activation = 'softmax'))

model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, 
         epochs=15, batch_size=100,
         validation_data=(x_test, y_test), verbose=2)
```
